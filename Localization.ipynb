```python
import numpy as np
import imageio
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.losses import binary_crossentropy, categorical_crossentropy
from tensorflow.keras import backend as K
from skimage.transform import resize
from glob import glob

from google.colab import drive
drive.mount('/content/drive/')
```

```python
images_folder = './drive/MyDrive/ColabNotebooks/image_localization/'
```

```python
# read images
ch = imageio.imread(images_folder+'poki_img/charmander-tight.png')
bb = imageio.imread(images_folder+'poki_img/bulbasaur-tight.png')
sq = imageio.imread(images_folder+'poki_img/squirtle-tight.png')
plt.imshow(sq)
plt.show()
```

```python
# save parameters of images
IMG_SIZE = 200
ch = np.array(ch)
bb = np.array(bb)
sq = np.array(sq)
CH_H, CH_W, CH_C = ch.shape
BB_H, BB_W, BB_C = bb.shape
SQ_H, SQ_W, SQ_C = sq.shape

poke_data = [
    [ch, CH_H, CH_W, CH_C],
    [bb, BB_H, BB_W, BB_C],
    [sq, SQ_H, SQ_W, SQ_C],
]
class_names = ['Charmander', 'Bulbasaur', 'Squirtle']
```

```python
# since we have several channels:
#   pokecmon presence, pokemon type, and its location
#   we note that in case of pokemon absence
#   loss of location and type should be neglected
#   thus we create corresponding loss funciton
def custom_loss(y_true, y_pred):

    bce = binary_crossentropy(y_true[:, :4], y_pred[:, :4])
    cce = categorical_crossentropy(y_true[:, 4:7], y_pred[:, 4:7])
    bce2 = binary_crossentropy(y_true[:, -1], y_pred[:, -1])

    return bce * y_true[:, -1] + cce * y_true[:, -1] + 0.5 * bce2
```

```python
# make a model
# including clearing the previouse session
def make_model():
    K.clear_session()
    vgg = tf.keras.applications.VGG16(
        input_shape=[IMG_SIZE, IMG_SIZE, 3],
        include_top=False,
        weights='imagenet')
    x = Flatten()(vgg.output)
    x1_prime = Dense(16, activation='ReLU')(x)
    x1 = Dense(4, activation='sigmoid')(x1_prime) # location
    x2_prime = Dense(12, activation='ReLU')(x)
    x2 = Dense(3, activation='softmax')(x2_prime) # object class
    x3_prime = Dense(4, activation='ReLU')(x)
    x3 = Dense(1, activation='sigmoid')(x3_prime) # object appeared
    x = Concatenate()([x1, x2, x3])
    model = Model(vgg.input, x)
    model.compile(loss=custom_loss, optimizer=Adam(lr=0.00005))
    return model
```

```python
backgrounds = []

background_files = glob(images_folder+'backgrounds_img/*.jpg')
for f in background_files:
    bg = np.array(image.load_img(f))
    backgrounds.append(bg)
```

```python
def pokemon_generator_multiclass(batch_size=64):
  
  while True:

    for _ in range(50):
        X = np.zeros((batch_size, IMG_SIZE, IMG_SIZE, 3))
        Y = np.zeros((batch_size, 8))

        for i in range(batch_size):

        bg_idx = np.random.choice(len(backgrounds))
        bg = backgrounds[bg_idx]
        bg_h, bg_w, _ = bg.shape
        rnd_h = np.random.randint(bg_h - IMG_SIZE)
        rnd_w = np.random.randint(bg_w - IMG_SIZE)
        X[i] = bg[rnd_h:rnd_h+IMG_SIZE,rnd_w:rnd_w+IMG_SIZE].copy()
        # randomle select one of the options:
        # i. no pokemon, ii. ch, iii. bb, iv. sq
        appear = (np.random.random() < 0.75)
        if appear:

            pk_idx = np.random.randint(3)
            pk, h, w, _ = poke_data[pk_idx]

            scale = 0.5 + np.random.random()
            new_height = int(h * scale)
            new_width = int(w * scale)
            obj = resize(
              pk,
              (new_height, new_width),
              preserve_range=True).astype(np.uint8)

            if np.random.random() < 0.5:
                obj = np.fliplr(obj)

            row0 = np.random.randint(IMG_SIZE - new_height)
            col0 = np.random.randint(IMG_SIZE - new_width)
            row1 = row0 + new_height
            col1 = col0 + new_width

            mask = (obj[:,:,3] == 0)
            bg_slice = X[i,row0:row1,col0:col1,:]
            bg_slice = np.expand_dims(mask, -1) * bg_slice
            bg_slice += obj[:,:,:3]
            X[i,row0:row1,col0:col1,:] = bg_slice
            # pokemon location
            Y[i,0] = row0/IMG_SIZE
            Y[i,1] = col0/IMG_SIZE
            Y[i,2] = (row1 - row0)/IMG_SIZE
            Y[i,3] = (col1 - col0)/IMG_SIZE
            # pokemon type
            Y[i,4 + pk_idx] = 1
        # appear on not
        Y[i,7] = appear

        yield X / 255., Y
```

```python
model = make_model()
model.fit_generator(
    pokemon_generator_multiclass(),
    steps_per_epoch=50,
    epochs=40,
)
```

```python
# Make predictions
def pokemon_prediction_multiclass():
  
    bg_idx = np.random.choice(len(backgrounds))
    bg = backgrounds[bg_idx]
    bg_h, bg_w, _ = bg.shape
    rnd_h = np.random.randint(bg_h - IMG_SIZE)
    rnd_w = np.random.randint(bg_w - IMG_SIZE)
    x = bg[rnd_h:rnd_h+IMG_SIZE,rnd_w:rnd_w+IMG_SIZE].copy()

    appear = (np.random.random() < 0.75)
    if appear:

        pk_idx = np.random.randint(3)
        pk, h, w, _ = poke_data[pk_idx]

        scale = 0.5 + np.random.random()
        new_height = int(h * scale)
        new_width = int(w * scale)
        obj = resize(
          pk,
          (new_height, new_width),
          preserve_range=True).astype(np.uint8)

    if np.random.random() < 0.5:
        obj = np.fliplr(obj)

    row0 = np.random.randint(IMG_SIZE - new_height)
    col0 = np.random.randint(IMG_SIZE - new_width)
    row1 = row0 + new_height
    col1 = col0 + new_width

    mask = (obj[:,:,3] == 0)
    bg_slice = x[row0:row1,col0:col1,:]
    bg_slice = np.expand_dims(mask, -1) * bg_slice
    bg_slice += obj[:,:,:3]
    x[row0:row1,col0:col1,:] = bg_slice
    actual_class = class_names[pk_idx]
    print("true:", row0, col0, row1, col1, actual_class)

    X = np.expand_dims(x, 0) / 255.
    p = model.predict(X)[0]

    fig, ax = plt.subplots(1)
    ax.imshow(x.astype(np.uint8))

    if p[-1] > 0.5:
        row0 = int(p[0]*IMG_SIZE)
        col0 = int(p[1]*IMG_SIZE)
        row1 = int(row0 + p[2]*IMG_SIZE)
        col1 = int(col0 + p[3]*IMG_SIZE)
        class_pred_idx = np.argmax(p[4:7])
        class_pred = class_names[class_pred_idx]
        print("pred:", row0, col0, row1, col1, class_pred)
        rect = Rectangle(
            (p[1]*IMG_SIZE, p[0]*IMG_SIZE),
            p[3]*IMG_SIZE, p[2]*IMG_SIZE,linewidth=1,edgecolor='r',facecolor='none')
        ax.add_patch(rect)
    else:
        print("pred: no object")
    plt.show()
```
